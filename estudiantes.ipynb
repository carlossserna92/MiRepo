```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
puntajeEstudiantes = pd.read_csv('studentscores.csv')
```

```python
puntajeEstudiantes.head(50)
```

```python
fig=plt.figure(figsize=(12,12))
plt.scatter(puntajeEstudiantes['Hours'], puntajeEstudiantes['Scores'])
plt.plot(puntajeEstudiantes['Hours'], puntajeEstudiantes['Scores'])
plt.xlabel('Horas dedicacion')
plt.ylabel('Calificaciones')
plt.grid()
```

```python
hdedicacion = puntajeEstudiantes['Hours'].values.reshape(-1,1)
calificaciones=puntajeEstudiantes['Scores'].values.reshape(-1,1)
linear_regressor = LinearRegression()
linear_regressor.fit(hdedicacion,calificaciones)
califPredecidas = linear_regressor.predict(hdedicacion)
califPredecidas
```

```python
#y=ax+b

a = linear_regressor.coef_[0][0]
b = linear_regressor.intercept_[0]
```

```python
fig=plt.figure(figsize=(12,12))
#plt.scatter(puntajeEstudiantes['Hours'], puntajeEstudiantes['Scores'])
plt.plot(puntajeEstudiantes['Hours'], puntajeEstudiantes['Scores'], label = 'Calificaciones medidas')
plt.plot(hdedicacion, califPredecidas, color = 'green', label='Calificaciones predecidas')
plt.xlabel('Horas dedicacion')
plt.ylabel('Calificaciones')
plt.legend()
plt.grid()
```
